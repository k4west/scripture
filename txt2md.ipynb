```python
def marp_script(title):
    marp_title = f'''---
marp: true
title: {title}
theme: gaia
class: lead
backgroundColor: #fff

'''
    return marp_title

css_style = '''style: |
  * {
    font-family: Arial, sans-serif;
    font-size: 19pt;
    line-height: 1.35em;
    color: black;
    }
  .scripture {
    text-align: left;
    text-align-last: left;
    padding-left: 1em;
    padding-right: 1.7em;
  }
  .scriptures {
    display: flex;
    flex-direction: column;
    justify-content: center;
    height: 100%;
  }
  .columns {
    display: grid;
    grid-template-columns: repeat(2, minmax(0, 1fr));
    gap: 1rem;
  }
  .image-container {
    display: flex;
    align-items: center;
    justify-content: center;
    height: 100%;
  }
  img {
    width: 550px;
    height: 600px;
    object-fit: contain;
  }

'''

def get_slide(kor1, eng1, kor2, eng2, idx):
  slide = f'''---

<div class="columns">
  <div class="scriptures">
    <br>
    <div class="scripture">
      <b>{kor1}
      </b>
    </div>
    <br>
    <div class="scripture">{eng1}
    </div>
    <br>
    <div class="scripture">
      <b>{kor2}
      </b>
    </div>
    <br>
    <div class="scripture">{eng2}
    </div>         
  </div>
  <div class="image-container">
    <img src='../../pictures/picture_{idx}.jpg'>
  </div>
</div>

'''
  return slide
```

```python
import os
import json
from random import sample


with open('JST_translation_2.json', 'r', encoding='utf-8') as f:
    JST_translation = json.load(f)

dst = 'slides'
for gospel in JST_translation:
    os.makedirs('/'.join((dst, gospel)), exist_ok=True)
li = [*range(179)]

for gospel in JST_translation:
    for chapter, verses in JST_translation[gospel].items():
        title = f'{gospel} {chapter}ìž¥'
        md_script = marp_script(title) + css_style
        contents = []
        content = []
        
        n = len(verses)
        n = n//2 + bool(n%2)
        
        for i in range(n):
            content.extend(verses[2*i:2*i+2])
            if len(''.join(content)) > 380:
                contents.append(content[:2] + ['', ''])
                content = content[2:]

            elif len(content) == 4:
                contents.append(content)
                content = []

        if content:
            if len(content) == 2:
                content.extend(['', ''])
            contents.append(content)

        pic = sample(li, len(contents))
        for i, (k1, e1, k2, e2) in enumerate(contents):
            md_script += get_slide(k1, e1, k2, e2, pic[i])
        with open(f'{dst}/{gospel}/{title}.md', 'w', encoding='utf-8') as f:
            f.write(md_script)

```
